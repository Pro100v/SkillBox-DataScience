```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('click_stream2.csv', header=None, names=['user_id', 'page', 'event_date', 'device'])
df['month'] = df['event_date'].str[:7]
```

```python
df.head()
```

```python
pvt = pd.pivot_table(df, values=['user_id'], index=['page'], columns=['device', 'month'], aggfunc=len)
pvt
```

```python
(pvt / pvt.iloc[0] * 100).round(2)
```
