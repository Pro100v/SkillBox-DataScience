```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('click_stream3.csv', header=None, names=['user_id', 'page', 'event_date', 'device', 'gender'])
df['month'] = df['event_date'].str[:7]
df.head()
```

```python
pvt = pd.pivot_table(df, values='user_id', index=['page'], columns=['device', 'gender', 'month',  ], aggfunc=len)
res = (pvt / pvt.iloc[0] * 100).round(2)
res
```

```python
res.loc['2_search_page':'3_payment_page', res.columns.get_level_values('month').isin(['2015-02', '2015-03'])]
```

Отбросив лишние страницы и месяцы (для простоты анализа), можно заметить, что влияние пола пользователей на ошибки возникшие в мобильной и десктопной версии приложения в марте месяце - отсутствует. При этом стоит отметить, что на десктопной версии страницы платежа, у мужчин падение более характерно чем у женщин: в 4.33 раза против 3.76 раза.
