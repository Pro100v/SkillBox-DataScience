```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
from matplotlib import colors as clr
```

```python
df = pd.read_csv('./data/AB_NYC_2019.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df[df['reviews_per_month'].isna()]['number_of_reviews'].sum()
```

```python
print('neighbourhood_group: ', df['neighbourhood_group'].unique())
print('neighbourhood: ', df['neighbourhood'].unique())
print('room_type: ', df['room_type'].unique())
```

```python
df.groupby(['room_type']).size()
```

```python
df.pivot_table(index=['room_type'], columns=['neighbourhood_group'], aggfunc=np.size, values='id')
```

```python
colors = clr.get_named_colors_mapping()
color_map = {
    'Brooklyn': colors['xkcd:cloudy blue'],
    'Manhattan': colors['xkcd:azul'],
    'Queens': colors['xkcd:electric lime'],
    'Staten Island': colors['xkcd:fresh green'],
    'Bronx': colors['xkcd:warm purple'],
}
```

```python
plt.figure(figsize=(10, 5))
# plt.imread('./data/New_York_City_.png')
for name, group in df.groupby(['neighbourhood_group']):
    plt.scatter(group['longitude'], group['latitude'], s=0.1, c=color_map[name], label=name)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.legend(markerscale=10)
plt.title('Accomodation location by neighbourhood_group')
```

## Bar

```python
plt.figure(figsize=(10, 5))
x = np.asarray([0, 1, 2])
width = 0.15
i = -2


# plt.imread('./data/New_York_City_.png')
for name, group in df.groupby(['neighbourhood_group']):
    plt.bar(x - i*width, group.groupby('room_type').size().values, width, label=name)
    i += 1
        
# plt.xlabel('longitude')
# plt.ylabel('latitude')
plt.legend(markerscale=10)
plt.title('Accomodation location by neighbourhood_group')
plt.xticks(x, list(df.groupby('room_type').groups.keys()))
```

## Hist

```python
plt.figure(figsize=(10, 5))
filtered_df = df[df['price'] < 1000]
for name, group in filtered_df.groupby(['neighbourhood_group']):
    plt.hist(group['price'], label=name, bins=50, alpha=0.3)
plt.xlabel('Price, $')
plt.ylabel('Accomodation quantity')
plt.title('Price distribution by neighbourhood_group')
plt.legend()
```

## Pie

```python
neighbourhood_group_size = df.groupby(['neighbourhood_group']).size()
neighbourhood_group_size
```

```python
plt.figure(figsize=(8, 8))
plt.pie(neighbourhood_group_size.values, labels=neighbourhood_group_size.index, autopct='%1.2f%%')
plt.legend()
```

## Boxplot

```python
filtered_df = df[df['price'] < 1000]
```

```python
plt.figure(figsize=(10,8))
index = 1
for name, group in filtered_df.groupby(['neighbourhood_group']):
    plt.boxplot(group['price'], positions=[index], labels=[name], autorange=False, showfliers=False)
    index += 1
plt.xlabel('neighbourhood_group')
plt.ylabel('Price, $')
plt.title('Price distribution by neighbourhood_group')
```
