### Задача 2

В одних координатных осях постройте линейные графики динамики продаж студии «Activision» в Северной Америке, Европе, Японии и во всем мире.

### Задача 3

В цикле (т.е. на разных координатных осях) постройте линейные графики динамики продаж студии «Activision» в Северной Америке, Европе, Японии и во всем мире.

```python
### YOUR CODE HERE ###
def build_graph(data, ax, parap):
    sns.lineplot(data=data, ax=ax, x='Year', y=param)
    ax.set_ylabel('')
    ax.set_xlabel('')
    ax.set_title(params[param])

params = {
    'NA_Sales':'Северная Америка', 
    'EU_Sales':'Европа',
    'JP_Sales':'Япония', 
    'Global_Sales':'Во всем мире'}

data = df[df['Publisher'].eq('Activision')]

# fig, axs = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(12, 3))
fig = plt.figure(tight_layout=True, figsize=(12, 8))
ax = None

for i, param in enumerate(list(params.keys())[:-1]):
    ax = plt.subplot2grid((2, 3), (0, i), fig=fig, sharey=ax)
    build_graph(data, ax, param)

param = list(params.keys())[-1]
ax = plt.subplot2grid((2,3), (1,0), colspan=3, fig=fig)
sns.lineplot(data=data, ax=ax, x='Year', y=param)
build_graph(data, ax, param)

fig.suptitle('Динамика средних продаж студии «Activision» по годам', size=14, weight='bold')
fig.text(.5, .01, 'Года', ha='center', weight='bold', size=12)
fig.text(-.01, .5, 'Объем продаж, млн. копий',  va='center', rotation='vertical', weight='bold', size=12)
# fig.tight_layout()
plt.show()
```

### Задача 5

### Задача 6

Постройте линейный график динамики количества игр, выпускаемых Nintendo, по годам. С помощью numpy или pandas определите промежуток, в который издатель каждый год выпускал больше 35 игр, и подсветите этот промежуток времени с помощью зелёного полупрозрачного прямоугольника.

```python
### YOUR CODE HERE ###
def get_time_intervals(data, date_field):
    start = stop = None
    time_intervals = []
    for index, row in data.iterrows():
        if not all([start, stop]):
            start = stop = row[date_field]
        else:
            if row[date_field] == stop + 1:
                stop = row[date_field]
            else:
                time_intervals.append((start, stop))
                start = stop = row[date_field]
    else:
        if all([start, stop]):
            time_intervals.append((start, stop))
    return time_intervals

def time_intervals_vspan(ax, data, clr, lbl=None):
    t_intervals = get_time_intervals(data, 'Year')
    for start, stop in t_intervals:
        ax.axvspan(start, stop, alpha=.25, color=clr, label=lbl)
    

data = df[df.Publisher.eq('Nintendo')].groupby('Year')['Name'].agg(GameCount=('count')).reset_index()

fig, ax = plt.subplots()
sns.lineplot(data=data, x='Year', y='GameCount', ax=ax)

time_intervals_vspan(
    ax=ax, clr='green', lbl='игр в год > 35',
    data=data[data['GameCount'] >=35])
time_intervals_vspan(
    ax=ax, clr='coral', lbl='игр в год < 10',
    data=data[data['GameCount'] < 10])

ax.set_title('Динамика количества игр, выпускаемых Nintendo, по годам', fontsize=14, weight='bold')
ax.set_xlabel('Года')
ax.set_ylabel('Кол-во игр в год')

# вывод уникальных значенией в легенде
# 
handles, labels = ax.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys())
ax.get_legend().set_title('Временные интервалы')

plt.show()
```

### Задача 7

Определите 3 жанра и 4 платформы с самыми большими продажами игр в мире за всё время. Постройте сетку графиков 4 на 3, в каждой ячейке постройте точечный график, проверяющий зависимость общемировых продаж от продаж в Северной Америке для каждой пары (жанра, платформы). Разными цветами укажите игры разных годов. Размер итогового полотна определите самостоятельно таким образом, чтобы все графики были видны.

Делайте графики достаточно читаемыми, чтобы из них можно было сделать вывод о наличии / отсутствии зависимостей в данных.

```python
top_platforms = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)[:4].index.to_list()
top_genre = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)[:3].index.to_list()

data = df[(df.Platform.isin(top_platforms)) & (df.Genre.isin(top_genre))]

g = sns.FacetGrid(
    data=data, 
    row='Genre', col='Platform',
    hue='Year', palette='crest'
)

g.map(sns.scatterplot, 'Global_Sales', 'NA_Sales',)
g.add_legend()
g.fig.suptitle('Соотношение продаж игр Общемировых к Североамериканиским\nмежду жанрами(ТОП-3) и платформами(ТОП-4)', 
    size=16, color='darkblue', weight='bold')
g.fig.subplots_adjust(top=.9)

for i, ax in enumerate(g.axes.flatten()):
    if i % 4 == 0:
        ax.set_ylabel('Северная Америка')
    if i // 4 == 2:
        ax.set_xlabel('Общемировые продажи')
    ax.set_title(
        ax.get_title().replace('Genre', 'Жанр').replace('Platform', 'Платформа')
    )

plt.show()
```
