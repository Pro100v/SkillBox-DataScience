## Набор данных от Kiva


### Описание данных
***

**kiva_loans** — это набор данных, который содержит большую часть информации о займах:
- id — уникальный идентификатор для кредита.
- funded_amount — сумма, выплаченная Kiva агенту на местах (USD).
- loan_amount — сумма, выплаченная полевым агентом заёмщику (USD).
- sector — сектор использования займа.
- activity — более гранулированная категория.
- use — назначение использования суммы кредита.
- country_code — ISO-код страны страны, в которой был выдан заём.
- country — полное название страны, в которой был выдан кредит.
- posted_time — время, в которое заявка была опубликована на Kiva.
- disbursed_time — время, когда кредитный агент выдал заёмщику сумму.
- funded_time — время, когда заявка полностью финансируется кредиторами.
- term_in_months — срок, на который был выдан кредит в месяцах.
- lender_count — общее количество кредиторов, которые внесли свой вклад в этот кредит.
- loaner_genders — разделённые запятыми буквы M, F, где каждый экземпляр представляет одного мужчину/женщину в группе.

**kiva_mpi_region_locations** — Набор данных о регионах с показателями MPI:
(Global Multidimensional Poverty Index — глобальный индекс бедности) ряда регионов (субнациональных) в мире.  Чем больше это значение, тем выше уровень бедности в соответствующем регионе. 
- country — страна.
- region — регион в стране.
- world_region — часть света.
- MPI — многомерный индекс бедности.
- geo — координаты (широта, долгота).

### Загрузка данных
 ***

```python
dt_kiva_path = './data/kiva_loans.csv' 
dt_MPI_path = './data/kiva_mpi_region_locations.csv'
load_rows = None
```

```python
# загрузка данных по заемам
df_kiva_loans = pd.read_csv(
    dt_kiva_path, 
    parse_dates=['posted_time', 'disbursed_time', 'funded_time', 'date'], 
    nrows=load_rows)
```

```python
# загрузка данных по макрорегионам
# после загрузки удалим столбец с геопозицеей и удалим все пустые строки (n/a)
df_mpi = pd.read_csv(dt_MPI_path, nrows=load_rows)\
    .drop(['geo'], axis=1)\
    .dropna(how='all')    
df_mpi.MPI.fillna(0, inplace=True)
```

```python
# Сгруппируем данные макрорегионов по странам и высчитаем медианное значение показателя бедности
df_mpi_country_grouped = df_mpi\
    .groupby(by=['world_region', 'ISO', 'country'])['MPI']\
    .median()\
    .fillna(0)\
    .reset_index()
```

## Обогащение набора данных дополнительными атрибутами

### Добавление информации по макрорегионам

```python
# расширим данные по займам информацией по макрорегионам и индексом бедности
df_kiva_loans = df_kiva_loans.merge(df_mpi_country_grouped, how='left', on='country')
```

### Расшифровка информации по гендерным группам

```python
import functools

@functools.lru_cache(None)
def gender_group(x: str) -> str: 
    '''
    функция для определения состава группы заемщиков
    return str:
        'male' - один заемщик мужчина 
        'female' - один заемщик женщина
        'male only' - группа заемщиков из одних мужчин
        'female only' - группа заемщиков из одних денщин
        'male great' - группа заемщиков, где большинство мужщин
        'female great' - группа заемщиков, где большинство женщин
        'm == f' - группа заемщиков с равным гендерным составом
        'undefine' - во всех остальных случаях
    '''
    undefine = 'undefine'
    if not isinstance(x, str):
        return undefine
    x = x.strip().lower()
    gen_set = set(x.split(', '))
    if x == 'male' or x == 'female':
        return x
    elif len(gen_set) == 1 and any(['male' in gen_set, 'female' in gen_set]):
        return 'male only' if 'male' in gen_set else 'female only'
    elif all([len(gen_set) == 2, 'male' in gen_set, 'female' in gen_set]):
        ser = pd.Series(x.split(', ')).value_counts()
        m, f = ser['male'], ser['female']
        return 'm == f' if m==f else 'male great' if m>f else 'female greate'
    else:
        return undefine
```

```python
# вычислим группы заемщиков
genders = {'male':'men', 'male only':'men', 'female':'women', 'female only':'women'}

df_kiva_loans['borrower_genders_group'] = df_kiva_loans['borrower_genders'].map(gender_group)

df_kiva_loans['borrower_genders_type'] = df_kiva_loans['borrower_genders_group'].map(genders).fillna('mix_groups')
```

### Классификация стран по кол-ву займов 
***

```python
# классифицируем страны на 3 группы по кол-ву займов
country_rate = pd.qcut(
    df_kiva_loans.country.value_counts(), 
    3, 
    labels=['low_country', 'middle_country', 'top_country'])
df_kiva_loans['country_rate'] = df_kiva_loans.country.map(country_rate)
```

### Расчет времени на сбор запрошенной суммы займа
***

```python
# drop na (missing) vlues
df_loans_date = df_kiva_loans.dropna(subset=['disbursed_time', 'funded_time'], how='any', inplace=False)

# dates columns
dates = ['posted_time', 'disbursed_time']

# format dates:
# df_loans_date[dates] = df_loans_date[dates].applymap(lambda x: x.split('+')[0])
# df_loans_date[dates] = df_loans_date[dates].apply(pd.to_datetime)

# calculate time interval 
df_kiva_loans.loc[:, 'time_funding'] = (df_loans_date[dates[0]] - df_loans_date[dates[1]]) / timedelta(days=1)
# df_kiva_loans.loc[:, 'time_funding'] = df_kiva_loans['time_funding'] / timedelta(days=1)
```

### Расчет ежемесячного платежа
***

```python
df_kiva_loans['monthly_repayment'] = df_kiva_loans['loan_amount'] / df_kiva_loans['term_in_months']
```

### Отделение от данных "выбросов" 
***

```python
# наблюдаемое отклонение от среднего
dev = (df_kiva_loans['loan_amount']-df_kiva_loans['loan_amount'].mean()).abs()
# Стандартное отклонение 
std = df_kiva_loans['loan_amount'].std()
# Фильтруем исходный набор данных 
df_kiva_loans_trimmed = df_kiva_loans[~(dev>3*std)]


# наблюдаемое отклонение от среднего
dev = (df_kiva_loans['loan_amount'] - df_kiva_loans['loan_amount'].mean()).abs()
# Стандартное отклонение 
std = df_kiva_loans['loan_amount'].std()
# набор данных без "выбросов"
df_kiva_loans_trimmed = df_kiva_loans[~(dev > 3*std)]
# набор данных состоящий из "выбросов"
df_kiva_loans_outlier = df_kiva_loans[(dev > 3*std)]
```

## Гендерный состав (п.2.2.2)
***
- В какких странах преобладают мужчины? 
- Группа людей? 

```python
# сформируем датасет сгруппированный по странам и типам заемщиков по их гендерному признаку и расчитаем количество заявок
#  после этого разложим тип заемщиков по разным столбцам

df_country_gender = df_kiva_loans\
    .groupby(['country', 'borrower_genders_type'])\
    ['id'].count()\
    .reset_index()\
    .pivot_table(index='country', columns='borrower_genders_type', values='id', aggfunc='sum')\
    .fillna(0)
# df_country_gender
```

```python
# опрежелим состав сторк в ТОП-Х
top = 5

def setup_ax(ax):
    ax.set_ylabel(None)
    ax.set_xlabel('Кол-во займов', fontsize=14)
    ax.legend(loc='lower right')
#     ax.tick_params(labelsize=16)

def get_df(x):
    return df_kiva_loans[df_kiva_loans.country.isin(x)]\
                .groupby(['country', 'borrower_genders_type'])\
                ['id'].count()\
                .reset_index()

fig, axs = plt.subplots(1, 2, figsize=(10, 0.8*top))
fig.suptitle(f'Распределение гендерных типов по странам (ТОП-{top})', fontsize=18, y=1.02)

# фильтр с преобладанием мужчин
select = (df_country_gender.men > df_country_gender.women) & (df_country_gender.men > df_country_gender.mix_groups)
df_tmp = df_country_gender[select].sort_values(by=['men'], ascending=False).head(top)

sns.barplot(y='country', x='id', hue = 'borrower_genders_type',
            data = get_df(df_tmp.index),
            order = df_tmp.index, alpha=.6, ax = axs[0])
setup_ax(axs[0])
axs[0].set_title('с преобладанием мужчин', fontsize=16)

# фильтр с преобладанием смешанных групп
select = (df_country_gender.mix_groups > df_country_gender.women) & (df_country_gender.mix_groups > df_country_gender.men)
df_tmp = df_country_gender[select].sort_values(by=['mix_groups'], ascending=False).head(top)

sns.barplot(y='country', x='id', hue = 'borrower_genders_type',
            data = get_df(df_tmp.index),
            order = df_tmp.index, alpha=.6, ax = axs[1])
setup_ax(axs[1])
axs[1].set_title('с преобладанием групп', fontsize=16)

fig.tight_layout()
plt.show()
```

## Анализ сумм заявок (п.2.4.3)
***
- Попытаться найти связь между странами (макрорегеонами), полом заемщика и суммой заявки.
- Какие выводы можно сделать?

```python
# Сформируем набор данных для анализа сумм заявок в разрезе макрорегионов и гендерных групп
df_loans_by_regions = \
    df_kiva_loans_trimmed\
    [df_kiva_loans_trimmed['borrower_genders_group'].ne('undefine')]\
    .groupby(by=['world_region', 'borrower_genders_group'])['loan_amount']\
    .agg(['median', 'sum'])\
    .reset_index()

# сформируем список для наглядной сортировки данных в графиках
gender_order = sorted(df_kiva_loans_trimmed['borrower_genders_group'].unique().tolist())
if 'undefine' in gender_order:
    gender_order.remove('undefine')
gender_order = gender_order[:4] + gender_order[-3:][::-1]
```

```python
figh, figw = 10, 5
cols = 2
rows = df_loans_by_regions['world_region'].unique().size 

xticks_medium = np.arange(0, max(df_loans_by_regions['median']+1), 500)
xticks_sum = np.arange(0, max(df_loans_by_regions['sum'])+10e6, 10e6)


fig, axs = plt.subplots(nrows=rows, ncols=cols, sharex=False, sharey=True, figsize=(figh*cols, figw*rows))
# axs = axs.flatten()

for i, wr in enumerate(df_loans_by_regions['world_region'].unique().tolist()):
    tmp_df = df_loans_by_regions[df_loans_by_regions.world_region.eq(wr)].dropna()
    
    sns.barplot(y='borrower_genders_group', x='median', order=gender_order, data=tmp_df, alpha=.6, ax=axs[i][0])
    
    axs[i][0].set_title(f"Медиана суммы займа в {wr} в разрезе пола", fontsize=18)
    axs[i][0].set_xlabel('Сумма займа в $ USA', fontsize=16)
    axs[i][0].set_xticks(xticks_medium)
    axs[i][0].set_ylabel(None)
    axs[i][0].tick_params(labelsize=14)
    
    sns.barplot(y='borrower_genders_group', x='sum', order=gender_order, data=tmp_df, alpha=.6, ax=axs[i][1])
    
    axs[i][1].set_title(f"общая сумма займов по {wr} в разрезе пола", fontsize=18)
    axs[i][1].set_xlabel('Сумма займа в $ USA', fontsize=16)
    axs[i][1].set_xticks(xticks_sum)
    axs[i][1].set_ylabel(None)
    axs[i][1].tick_params(labelsize=14)
        
#     ax.barh(tmp_df.activity, tmp_df.loan_amount, align='center')
#     ax.tick_params(axis='both', which='major', labelsize=18)
    
#     ax.set_title(q, fontsize=20)
#     ax.yaxis.label.set_visible(False)    
plt.tight_layout()
plt.show()
```

Приведенные сравниетельные графики по показателю **медианы суммы займа** в разрезе мировых регионов и гендерной принадлежности заемщиков показывает, что займы свыше **$1000** женщинами практически не запрашиваются или толькр в группе заемшиков. Мужчины чаще запрашивают займы с крупной суммой займа. 

Если рассматривать **общие суммы займов** за весь изучаемый период, то можно заметить, что в **Европейском регионе**, **Центральной** и **Южной Азии** практически отсутствуют займы в группе заемщиков, в них привалирует индивидуальные займы.

У мужчин, по соседству с последними двумя секторамами находятся *Образование* и *Сфера услуг* . К отусайдерам же можно отнести *Оптовые продажи* и *Сферу развлечения*

## Влияние показателя **MPI** (*многомерный индекс бедности*) в разных странах на суммы займов и сроки погашения

```python
mpi_rt_labela = ['низкий','ниже среднего', 'средний', 'выше среднего', 'высокий']

df_kiva_loans['MPI_rate'] =  pd.cut(df_kiva_loans['MPI'], bins=len(mpi_rt_labela), labels=mpi_rt_labela)
# df_kiva_loans['MPI_rate'] = df_kiva_loans['MPI_rate'].cat.add_categories('нет данных')
# df_kiva_loans['MPI_rate'].fillna('нет данных', inplace=True)
```

```python
df_tmp[df_tmp > 0]
```

```python
df_tmp = df_kiva_loans[~df_kiva_loans.MPI.isna()].MPI_rate.value_counts()

# отбросим пустые данные
df_tmp = df_tmp[df_tmp > 0]

# df_tmp = df_kiva_loans[~df_kiva_loans.MPI.isna()].MPI_rate.value_counts()
fig, axs = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Распределение индекса бедности, MPI', fontsize=18)

sns.distplot(df_kiva_loans.MPI, ax=axs[0])
axs[0].set_xlabel('Натуральное распределение', fontsize=14)
axs[0].set_ylabel('Доля займов', fontsize=12)

sns.barplot(x=df_tmp.index, y=df_tmp.values, ax=axs[1], alpha=0.6)
axs[1].set_xlabel('Распределение по группам', fontsize=14)
axs[1].set_ylabel('Кол-во займов', fontsize=12)

plt.show()
```

Общая картина по коэффициенту бедности показывает, что количество займов пропорционально уменьшается с повышением уровнем бедности в странах. Это возможно связанно с тем, что в бедных странах хуже с доступом в интернет и в частности доступ к площадке Kiva, поэтому стоит подумать о повышении доступности площадки для слоев населения с отсутствием доступа к интернету

```python
# рассмотрим только строки с не пустые индексом бедности
mask = ~df_kiva_loans.MPI.isna()

df_tmp = df_kiva_loans[mask]\
    .groupby(['world_region', 'MPI_rate'])\
    .agg({'loan_amount': 'median', 'term_in_months':'mean', })\
    .reset_index()

# подготовим датасет для одновременного вывода на графике данных по суммам займов и срокам
df_x = df_tmp.melt(id_vars=['world_region', 'MPI_rate'], var_name="source", value_name="value_numbers")
mask = df_x.source.eq('loan_amount')

# рассчитаем коэф-нт масштабирования
scale = (df_x[mask].value_numbers.mean()
         / df_x[~mask].value_numbers.mean())

df_x['value_numbers_scale'] = df_x['value_numbers']
df_x.loc[~mask, 'value_numbers_scale'] =  df_x.loc[~mask, 'value_numbers'] * scale
df_x.sample(10)
```

```python
g = sns.catplot(
    y='value_numbers_scale', x='MPI_rate', hue='source', data=df_x, 
    col='world_region', col_wrap=2, kind='bar', 
    sharey=False, sharex=False, alpha=0.6)

(g.set_axis_labels("Индекс бедности", "Медиана суммы займа, $ USD")
    .set_titles("Макрорегион: {col_name} ")
    .despine(left=True)) 
g.fig.suptitle('Распределение суммы и сроков займов\n в разрезе макрорегионов и групп индекса бедности', 
               fontsize=18, y=1.05)


# 
for i, ax in enumerate(g.axes):
    ax2 = ax.twinx()
    
    ax2.set_ylim(ax.get_ylim())
    ax2.set_yticklabels(np.round(ax.get_yticks()/scale,0).astype(int))    
    ax2.set_ylabel('' if i%2 ==0 else 'Среднего срока займа, мес')

plt.tight_layout()
plt.show()
```

можно отметить Восточную Азию и Тихоокеанский макрорегион, в котором в отличии от остальных макрорегионов с повышением индекса бедности так же растут и медианные суммы займов - в других картина не так однозначна. 

Сроки займов с повышением индекса бедности чаще уменьшаются, особенно показателен разрыв между *низким* индексом бедности и *ниже средним*.

## Зависимость таких метрик как *Сумма займа*, *Срока займа*, *Время финансирования заявки*, *Ежемесячный платеж* в разрезе **Макрорегионов**

```python
df_tmp = df_kiva_loans\
    .groupby(by='world_region')\
    .agg(
        {
        'loan_amount': 'median',
        'monthly_repayment': 'median',
        'term_in_months': 'mean',
        'time_funding': 'mean',        
        })\
    .reset_index()\
    .melt(id_vars='world_region', var_name="source", value_name="value_numbers")

translate_dict = {
    'loan_amount': 'Сумма займа (медиана), $USD',
    'monthly_repayment': 'Ежемесячный платеж (медиана), $USD',
    'term_in_months': 'Срок займа (среднее), мес.',
    'time_funding': 'Время финансирования (среднее), дни', 
}

df_tmp.source = df_tmp.source.map(translate_dict)

g = sns.catplot(x='value_numbers', y='world_region', data=df_tmp, 
            col='source', col_wrap=2,
            kind='bar', sharex=False, 
            palette = 'inferno', alpha=.65,
            height=4, aspect=1.5, ci=None,  )

(g.set_axis_labels("", "")
    .set_titles("Метрика: {col_name} ", size=14)
    .despine(left=True)) 
g.fig.suptitle('Зависимость различных метрик займа в разрезе Макрорегионов', 
               fontsize=18, y=1.05)

plt.show()
```

* На представленных графиках **с положительной стороны** можно отметить *Южно-Африканский регион*, т.к. в нем самое минимальное среднее время сбора запрашиваемых займов. С учетом предыдущего графика, где демография стран с различными группами индекса бедности представлена широко, можно с уверенностью говорить об успешной активности Kiva в этомм регионе. 

* С другой стороны *Востояная Азия и Тихоокеанский реион*, где отражена самые большие сроки на сбор средств, при этом запрашиваемые суммы займа самые маленькие по сравнению с другими регионами. На предыдущих графиках так же видно, что повышеные суммы займов запрашивают группы, где индекс бедности выше среднего. Данный регион требует наиболее пристольного внимания для продвижения площадки Kiva, а так же продвигать заявки на сбор средств у кредиторов.
